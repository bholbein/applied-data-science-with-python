```python
import pandas as pd

dataframe = pd.read_csv('kicker2018.csv', decimal=",")
```

```python
dataframe
```

```python
columns_to_keep = ['Unnamed: 1',
                   'Verein',
                   'Pos.',
                   'Mio.',
                   'PI V.',
                   'DN V.']
kicker = dataframe[columns_to_keep]
kicker.head()
```

```python
kicker = kicker.rename(columns={'Unnamed: 1': 'Spieler', 'Pos.': 'Position', 'Mio.': 'Wert Kicker', 'PI V.': 'Punkte Vorsaison', 'DN V.': 'Durchschnittsnote Vorsaison'})
kicker.head()
```

```python
kicker['Punkte pro Wert'] = kicker['Punkte Vorsaison'] / kicker['Wert Kicker']
```

```python
kicker.head()
```

```python
kicker['Punkte pro Wert'] = kicker['Punkte pro Wert'].round(decimals=-1)
```

```python
kicker.head()
```

```python
kicker.sort_values(by=['Position', 'Punkte pro Wert', 'Durchschnittsnote Vorsaison', 'Wert Kicker'], ascending=[False, False, True, True])
```

```python
kicker_tor = kicker[kicker['Position'] == 'TOR']
kicker_abw = kicker[kicker['Position'] == 'ABW']
kicker_mit = kicker[kicker['Position'] == 'MIT']
kicker_stu = kicker[kicker['Position'] == 'STU']
```
